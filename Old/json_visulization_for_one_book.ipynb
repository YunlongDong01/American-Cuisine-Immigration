```python
import json
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import pandas as pd

# 1. 加载数据
with open('miss_extracted.json', 'r', encoding='utf-8') as f:
    cookbook = json.load(f)

recipes = cookbook['recipes']
print(f"总菜谱数量: {len(recipes)}")

# 2. 数据准备
df = pd.DataFrame(recipes)

# 分析1：各ethnicgroup的菜谱数量柱状图
plt.figure(figsize=(12, 6))
ethnic_counts = df['ethnic_group'].value_counts()

# 处理空值
ethnic_counts = ethnic_counts[ethnic_counts.index != '']
if ethnic_counts.empty:
    ethnic_counts['unknown'] = len(df)
    print("未找到明确的ethnic_group分类")

ax = ethnic_counts.plot(kind='bar', color='skyblue')
plt.title(f"《{cookbook['metadata']['title']}》中各民族的菜谱数量", fontsize=14)
plt.xlabel('民族', fontsize=12)
plt.ylabel('菜谱数量', fontsize=12)
plt.xticks(rotation=45)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.tight_layout()
plt.show()
```

```python
# 分析2：各ethnicgroup的食材词云
def generate_wordcloud(ingredient_list, title):
    all_ingredients = [ing for sublist in ingredient_list for ing in sublist]
    if not all_ingredients:
        print(f"{title} - 无食材数据")
        return
    
    # 计算食材频率
    freq = Counter(all_ingredients)
    
    # 生成词云
    wc = WordCloud(width=800, height=400, 
                  background_color='white',
                  colormap='tab20',
                  max_words=50).generate_from_frequencies(freq)
    
    plt.figure(figsize=(12, 6))
    plt.imshow(wc, interpolation='bilinear')
    plt.title(f"{title}常用食材", fontsize=14)
    plt.axis('off')
    plt.show()
    
    # 显示前10高频食材
    print("高频食材TOP10:")
    for item, count in freq.most_common(10):
        print(f"{item}: {count}次")

# 按ethnicgroup分组展示
if 'ethnic_group' in df.columns:
    grouped = df.groupby('ethnic_group')
    for name, group in grouped:
        if name == '':  # 跳过空分类
            continue
        print(f"\n民族分类: {name.upper()} (共{len(group)}道菜)")
        generate_wordcloud(group['ingredients'].tolist(), f"{name}民族")
else:
    print("数据中未找到ethnic_group字段")
    generate_wordcloud(df['ingredients'].tolist(), "全部菜谱")

# 分析3：各class1的分布和食材特征
plt.figure(figsize=(12, 6))
class_counts = df['class1'].value_counts()

ax = class_counts.plot(kind='barh', color='lightgreen')
plt.title(f"《{cookbook['metadata']['title']}》中各类别菜谱数量", fontsize=14)
plt.xlabel('菜谱数量', fontsize=12)
plt.ylabel('类别', fontsize=12)
for p in ax.patches:
    ax.annotate(str(p.get_width()), (p.get_width(), p.get_y() + p.get_height() / 2.),
                ha='left', va='center', xytext=(5, 0), textcoords='offset points')
plt.tight_layout()
plt.show()

# 各类别的食材特征
if 'class1' in df.columns:
    grouped_class = df.groupby('class1')
    for name, group in grouped_class:
        print(f"\n类别: {name.upper()} (共{len(group)}道菜)")
        generate_wordcloud(group['ingredients'].tolist(), f"{name}类别")
else:
    print("数据中未找到class1字段")
```
